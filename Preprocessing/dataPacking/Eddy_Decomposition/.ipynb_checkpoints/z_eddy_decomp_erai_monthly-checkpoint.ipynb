Description  : This script serves to decompose the eddy components of the geopotential at 500hPa based on subdaily data. The eddies of Z500 is a good indicator of the behavior of cyclones and many behaviors at midlatitudes. This can provide extra information about the role of transient eddy and stationary eddy when we investigate the meridional energy transport.<br>

Return Values : netCDF4 <br>
Caveat!       :  The data used here across subdaily scale. The saving freqeuncy of data is 6 hourly.<br>
The procedure is generic and is able to adapt any atmospheric reanalysis datasets, with some changes. Referring to the book "Physics of Climate", the concept of decomposition of circulation is given with full details. As a consequence, the meridional energy transport can be decomposed into 4 parts:<br>
**A = \[/overbar{A}\] + /ovrebar{A*} + \[A\]' + A'* ** <br>

- \[/overbar{A}\]:    energy transport by steady mean circulation<br>
- /ovrebar{A*}:     energy transport by stationary eddy<br>
- \[A\]':             energy transport by transient eddy<br>
- A'*:              energy transport by instantaneous and asymmetric part<br>

An example is given at page 277, in terms of transport of moisture.<br>
Here we will focus on three components of total meridional energy transport:<br>
** \[/overbar{vT}\] = \[/overbar{v}\] x \[/overbar{T}\] + \[/overbar{v}* x /overbar{T}*\] + /overbar{v'T'}\] ** <br>

- \[/overbar{v}\] x \[/overbar{T}\]:energy transport by steady mean circulation <br>
- \[/overbar{v}* x /overbar{T}*\]:    energy transport by stationary eddy <br>
- \[/overbar{v'T'}\]:                 energy transport by transient eddy <br>

```python
##########################################################################
###########################   Units vacabulory   #########################
# cpT:  [J / kg K] * [K]     = [J / kg]
# Lvq:  [J / kg] * [kg / kg] = [J / kg]
# gz in [m2 / s2] = [ kg m2 / kg s2 ] = [J / kg]

# multiply by v: [J / kg] * [m / s] => [J m / kg s]
# sum over longitudes [J m / kg s] * [ m ] = [J m2 / kg s]

# integrate over pressure: dp: [Pa] = [N m-2] = [kg m2 s-2 m-2] = [kg s-2]
# [J m2 / kg s] * [Pa] = [J m2 / kg s] * [kg / s2] = [J m2 / s3]
# and factor 1/g: [J m2 / s3] * [s2 /m2] = [J / s] = [Wat]
##########################################################################
```

```python
# constants
constant = {'g' : 9.80616,      # gravititional acceleration [m / s2]
            'R' : 6371009,      # radius of the earth [m]
            'cp': 1004.64,      # heat capacity of air [J/(Kg*K)]
            'Lv': 2264670,      # Latent heat of vaporization [J/Kg]
            'R_dry' : 286.9,    # gas constant of dry air [J/(kg*K)]
            'R_vap' : 461.5,    # gas constant for water vapour [J/(kg*K)]
            }
```

```python
################################   Input zone  ######################################
# specify starting and ending time
start_year = 1979
end_year = 2017
# choose the slice number for the vertical layer
lev_slice = 1
# specify data path
# ERAI 3D fields on pressure level
datapath = '/home/ESLT0068/WorkFlow/Core_Database_AMET_OMET_reanalysis/ERAI/regression/pressure/daily'
# specify output path for figures
output_path = '/home/ESLT0068/WorkFlow/Core_Database_AMET_OMET_reanalysis/ERAI/regression'
# benchmark datasets for basic dimensions
benchmark_file = 'pressure_daily_075_diagnostic_1998_3_z.nc'
benchmark = Dataset(os.path.join(datapath, 'era1998', benchmark_file))
####################################################################################
```

```python
def var_key_retrieve(datapath, year, month):
    # get the path to each datasets
    print ("Start retrieving datasets {} (y) {} (m)".format(year,month))
    # The shape of each variable is (241,480)
    datapath = os.path.join(datapath, 'era{}'.format(year),
                            'pressure_daily_075_diagnostic_{}_{}_z.nc'.format(year,month))
    # get the variable keys
    var_key = Dataset(datapath)

    print ("Retrieving datasets successfully and return the variable key!")
    return var_key
```

```python
def initialization(benchmark):
    print ("Prepare for the main work!")
    # create the month index
    period = np.arange(start_year,end_year+1,1)
    index_month = np.arange(1,13,1)
    # create dimensions for saving data
    #Dim_level = len(benchmark.variables['level'][:])
    Dim_latitude = len(benchmark.variables['latitude'][:])
    Dim_longitude = len(benchmark.variables['longitude'][:])
    Dim_month = len(index_month)
    Dim_period = len(period)
    #latitude = benchmark.variables['latitude'][:]
    #longitude = benchmark.variables['longitude'][:]
    #Dim_time = len(benchmark.variables['time'][:])
    # a list of the index of starting day in each month
    month_day_length = [31,28,31,30,31,30,31,31,30,31,30,31] #! we ignore the last day of February for the leap year
    month_day_index = [0,31,59,90,120,151,181,212,243,273,304,334]
    # create variables
    var_temporal_sum = np.zeros((365,Dim_latitude,Dim_longitude),dtype=float) #! we ignore the last day of February for the leap year
    return period, index_month, Dim_latitude, Dim_longitude, Dim_month, Dim_period,\
           month_day_length, month_day_index, var_temporal_sum
```

```python
def pick_var(var_key):
    # validate time and location info
    time = var_key.variables['time'][:]
    level = var_key.variables['level'][:]
    latitude = var_key.variables['latitude'][:]
    longitude = var_key.variables['longitude'][:]
    date = num2date(time,var_key.variables['time'].units)
    days = len(time)//4
    print ('*******************************************************************')
    print ('The datasets contain information from %s to %s' % (date[0],date[-1]))
    print ('There are %d days in this month' % (len(time)//4))
    print ('The coordinates include %d vertical levels' % (len(level)))
    print ('The grid employs %d points in latitude, and %d points in longitude' % (len(latitude),len(longitude)))
    print ('*******************************************************************')
    # extract variables
    print ("Start extracting velocity for the calculation of mean over time and space.")
    # extract data at certain levels
    var = np.zeros((len(time),len(latitude),len(longitude)),dtype=float)
    var[:,:,:] = var_key.variables['z'][:,lev_slice,:,:] / constant['g']
    # daily mean
    # first we reshape the array
    var_expand = var.reshape(len(time)//4,4,len(latitude),len(longitude))
    # Then we take daily mean
    var_daily = np.mean(var_expand,1)
    if days == 29:
        var_out = var_daily[:-1,:,:]
    else:
        var_out = var_daily
    print ('Extracting variables successfully!')

    return var_out
```

```python
def initialization_eddy(var_temporal_mean):
    # Here we only use the temporal mean, for the spatial mean we will take it dynamically
    # during the calculation of eddies, for the sake of memory usage.
    # create space for eddies
    var_2_transient_pool = np.zeros((Dim_period,Dim_month,Dim_latitude,Dim_longitude),dtype=float)
    var_2_standing_pool = np.zeros((Dim_period,Dim_month,Dim_latitude,Dim_longitude),dtype=float)
    var_2_transient_mean_pool = np.zeros((Dim_period,Dim_month,Dim_latitude),dtype=float)
    var_2_stationary_mean_pool = np.zeros((Dim_period,Dim_month,Dim_latitude,Dim_longitude),dtype=float)
    # create space for overall momentum
    var_2_overall_pool = np.zeros((Dim_period,Dim_month,Dim_latitude,Dim_longitude),dtype=float)
    # calculate mean meridional circulation
    var_2_steady_mean_zonal_mean = np.mean(var_temporal_mean,2)
    var_2_steady_mean_monthly_zonal_mean = np.zeros((12,Dim_latitude),dtype=float)
    for i in np.arange(Dim_month):
        var_2_steady_mean_monthly_zonal_mean[i,:] = np.mean(var_2_steady_mean_zonal_mean[month_day_index[i-1]:month_day_index[i-1]+month_day_length[i-1],:],0)
    var_2_steady_mean = var_2_steady_mean_monthly_zonal_mean * var_2_steady_mean_monthly_zonal_mean

    return var_2_transient_pool, var_2_standing_pool, var_2_transient_mean_pool,\
           var_2_stationary_mean_pool, var_2_overall_pool, var_2_steady_mean
```

```python
def compute_eddy(var_temporal_mean_select, var):
    '''
    We follow the method given by Peixoto and Oort, 1983.
    The equation is listed on page 61-63.
    equation 4.6 and 4.10
    The example is given on page 288.
    Here we take our naming convention for different eddies.
    For the details, please visit "Transient & Standing eddy"
    in notes.
    '''
    # shape of v[days,lat,lon]
    # calculate transient eddies
    ################# transient eddy ###################
    print ("Calculate transient eddies!")
    var_prime = var - var_temporal_mean_select
    var_2_transient = var_prime * var_prime
    # monthly mean
    # shape[lat,lon]
    var_2_transient_monthly_mean = np.mean(var_2_transient,0)
    ####################################################
    # calculate transient mean eddies
    ############### transient mean eddy ################
    print ("Calculate transient mean eddies!")
    var_prime_zonal_mean = np.mean(var,2) - np.mean(var_temporal_mean_select,2)
    var_2_transient_mean = var_prime_zonal_mean * var_prime_zonal_mean
    # monthly mean
    # shape[lat]
    var_2_transient_mean_monthly_mean = np.mean(var_2_transient_mean,0)
    ####################################################
    # Calculate standing eddies
    ################## standing eddy ###################
    print ("Calculate standing eddies!")
    var_star = np.zeros(var.shape,dtype=float)
    var_zonal_mean = np.mean(var,2)
    var_zonal_mean_enlarge = np.repeat(var_zonal_mean[:,:,np.newaxis],Dim_longitude,2)
    var_star = var - var_zonal_mean_enlarge
    var_2_standing = var_star * var_star
    # monthly mean
    # shape[lat,lon]
    var_2_standing_monthly_mean = np.mean(var_2_standing,0)
    ####################################################
    # Calculate stationary mean eddies
    ##############  stationary mean eddy ###############
    print ("Calculate stationary mean eddies!")
    var_monthly_mean = np.mean(var,0)
    var_monthly_zonal_mean = np.mean(var_monthly_mean,1)
    var_monthly_zonal_mean_enlarge = np.repeat(var_monthly_zonal_mean[:,np.newaxis],Dim_longitude,1)
    var_star_monthly_zonal_mean = var_monthly_mean - var_monthly_zonal_mean_enlarge
    # monthly mean
    # shape[lat,lon]
    var_2_stationary_mean_monthly_mean = var_star_monthly_zonal_mean * var_star_monthly_zonal_mean
    ####################################################
    # calculate the overall momentum transport
    ##############   overall transport   ###############
    print ("Calculate overall momentum transport!")
    var_2_overall = var * var
    # monthly mean
    # shape[lat,lon]
    var_2_overall_monthly_mean = np.mean(var_2_overall,0)
    ####################################################

    return var_2_transient_monthly_mean, var_2_transient_mean_monthly_mean,\
           var_2_standing_monthly_mean, var_2_stationary_mean_monthly_mean,\
           var_2_overall_monthly_mean
```

```python
def create_netcdf_point_eddy(var2_overall,var2_transient,var2_transient_mean,var2_standing,
                             var2_stationary_mean,var2_steady_mean,output_path):
    # take the zonal mean
    var2_overall_zonal = np.mean(var2_overall,3)
    var2_transient_zonal = np.mean(var2_transient,3)
    # v2_transient_mean is zonal mean already
    var2_standing_zonal = np.mean(var2_standing,3)
    var2_stationary_mean_zonal = np.mean(var2_stationary_mean,3)
    # create netCDF
    print ('*******************************************************************')
    print ('*********************** create netcdf file*************************')
    print ('*******************************************************************')
    # wrap the datasets into netcdf file
    # 'NETCDF3_CLASSIC', 'NETCDF3_64BIT', 'NETCDF4_CLASSIC', and 'NETCDF4'
    data_wrap = Dataset(os.path.join(output_path,'model_daily_075_z500_eddies_point.nc'),'w',format = 'NETCDF4')
    # create dimensions for netcdf data
    year_wrap_dim = data_wrap.createDimension('year',Dim_period)
    month_wrap_dim = data_wrap.createDimension('month',Dim_month)
    lat_wrap_dim = data_wrap.createDimension('latitude',Dim_latitude)
    lon_wrap_dim = data_wrap.createDimension('longitude',Dim_longitude)
    # create coordinate variables for 1-dimensions
    year_wrap_var = data_wrap.createVariable('year',np.int32,('year',))
    month_wrap_var = data_wrap.createVariable('month',np.int32,('month',))
    lat_wrap_var = data_wrap.createVariable('latitude',np.float32,('latitude',))
    lon_wrap_var = data_wrap.createVariable('longitude',np.float32,('longitude',))
    # create the 4-d variable
    var2_overall_wrap_var = data_wrap.createVariable('overall',np.float64,('year','month','latitude','longitude'),zlib=True)
    var2_transient_wrap_var = data_wrap.createVariable('transient',np.float64,('year','month','latitude','longitude'),zlib=True)
    var2_standing_wrap_var = data_wrap.createVariable('standing',np.float64,('year','month','latitude','longitude'),zlib=True)
    var2_stationary_mean_wrap_var = data_wrap.createVariable('stationary_mean',np.float64,('year','month','latitude','longitude'),zlib=True)
    # create the 4d variable
    var2_overall_zonal_wrap_var = data_wrap.createVariable('overall_zonal',np.float64,('year','month','latitude'),zlib=True)
    var2_transient_zonal_wrap_var = data_wrap.createVariable('transient_zonal',np.float64,('year','month','latitude'),zlib=True)
    var2_transient_mean_wrap_var = data_wrap.createVariable('transient_mean',np.float64,('year','month','latitude'),zlib=True)
    var2_standing_zonal_wrap_var = data_wrap.createVariable('standing_zonal',np.float64,('year','month','latitude'),zlib=True)
    var2_stationary_mean_zonal_wrap_var = data_wrap.createVariable('stationary_mean_zonal',np.float64,('year','month','latitude'),zlib=True)
    # create the 2d variable
    var2_steady_mean_wrap_var = data_wrap.createVariable('steady_mean',np.float64,('month','latitude'),zlib=True)
    # global attributes
    data_wrap.description = 'Monthly stationary and transient eddies at each grid point'
    # variable attributes
    lat_wrap_var.units = 'degree_north'
    lon_wrap_var.units = 'degree_east'
    var2_overall_wrap_var.units = 'm2'
    var2_transient_wrap_var.units = 'm2'
    var2_standing_wrap_var.units = 'm2'
    var2_stationary_mean_wrap_var.units = 'm2'
    var2_overall_zonal_wrap_var.units = 'm2'
    var2_transient_zonal_wrap_var.units = 'm2'
    var2_transient_mean_wrap_var.units = 'm2'
    var2_standing_zonal_wrap_var.units = 'm2'
    var2_stationary_mean_zonal_wrap_var.units = 'm2'
    var2_steady_mean_wrap_var.units = 'm2'

    lat_wrap_var.long_name = 'Latitude'
    lon_wrap_var.long_name = 'Longitude'
    var2_overall_wrap_var.long_name = 'Northward transport of momentum by all motions'
    var2_transient_wrap_var.long_name = 'Northward transport of momentum by transient eddy'
    var2_standing_wrap_var.long_name = 'Northward transport of momentum by standing eddy'
    var2_stationary_mean_wrap_var.long_name = 'Northward transport of momentum by stationary mean eddy'
    var2_overall_zonal_wrap_var.long_name = 'Zonal mean of northward transport of momentum by all motions'
    var2_transient_zonal_wrap_var.long_name = 'Zonal mean of northward transport of momentum by transient eddy'
    var2_transient_mean_wrap_var.long_name = 'Northward transport of momentum by transient mean eddy'
    var2_standing_zonal_wrap_var.long_name = 'Zonal mean of northward transport of momentum by standing eddy'
    var2_stationary_mean_zonal_wrap_var.long_name = 'Zonal mean of northward transport of momentum by stationary mean eddy'
    var2_steady_mean_wrap_var.long_name = 'Northward transport of momentum by steady mean meridional circulation'

    # writing data
    year_wrap_var[:] = period
    month_wrap_var[:] = index_month
    lat_wrap_var[:] = benchmark.variables['latitude'][:]
    lon_wrap_var[:] = benchmark.variables['longitude'][:]
    var2_overall_wrap_var[:] = var2_overall
    var2_transient_wrap_var[:] = var2_transient
    var2_standing_wrap_var[:] = var2_standing
    var2_stationary_mean_wrap_var[:] = var2_stationary_mean
    var2_overall_zonal_wrap_var[:] = var2_overall_zonal
    var2_transient_zonal_wrap_var[:] = var2_transient_zonal
    var2_transient_mean_wrap_var[:] = var2_transient_mean
    var2_standing_zonal_wrap_var[:] = var2_standing_zonal
    var2_stationary_mean_zonal_wrap_var[:] = var2_stationary_mean_zonal
    var2_steady_mean_wrap_var[:] = var2_steady_mean

    # close the file
    data_wrap.close()
    print ("The generation of netcdf files for fields on surface is complete!!")
```

```python
if __name__=="__main__":
    # calculate the time for the code execution
    start_time = tttt.time()
    # initialization
    period, index_month, Dim_latitude, Dim_longitude, Dim_month, Dim_period,\
    month_day_length, month_day_index, var_temporal_sum = initialization(benchmark)
    print ('*******************************************************************')
    print ('************  calculate the temporal and spatial mean  ************')
    print ('*******************************************************************')
    for i in period:
        for j in index_month:
            # get the key of each variable
            variable_key = var_key_retrieve(datapath,i,j)
            # take the daily mean of target fields
            variable = pick_var(variable_key)
            # add daily field to the summation operator
            var_temporal_sum[month_day_index[j-1]:month_day_index[j-1]+month_day_length[j-1],:,:] = \
            var_temporal_sum[month_day_index[j-1]:month_day_index[j-1]+month_day_length[j-1],:,:] + variable
    # calculate the temporal mean
    var_temporal_mean = var_temporal_sum / Dim_period
    print ('*******************************************************************')
    print ('**********  calculate the stationary and transient eddy  **********')
    print ('*******************************************************************')
    # Initialization
    # Grab temporal & spatial mean
    # The mean meridional circulation is calculated here
    var_2_transient_pool, var_2_standing_pool, var_2_transient_mean_pool,\
    var_2_stationary_mean_pool, var_2_overall_pool, var_2_steady_mean = initialization_eddy(var_temporal_mean)
    for i in period:
        for j in index_month:
            # get the key of each variable
            variable_key = var_key_retrieve(datapath,i,j)
            # take the daily mean of target fields at certain levels
            variable = pick_var(variable_key)
            var_temporal_mean_select = var_temporal_mean[month_day_index[j-1]:month_day_index[j-1]+month_day_length[j-1],:,:]
            var_2_transient, var_2_transient_mean, var_2_standing, var_2_stationary_mean,\
            var_2_overall = compute_eddy(var_temporal_mean_select, variable)
            # save output to the data pool for netCDF
            var_2_overall_pool[i-start_year,j-1,:,:] = var_2_overall
            var_2_transient_pool[i-start_year,j-1,:,:] = var_2_transient
            var_2_transient_mean_pool[i-start_year,j-1,:] = var_2_transient_mean
            var_2_standing_pool[i-start_year,j-1,:,:] = var_2_standing
            var_2_stationary_mean_pool[i-start_year,j-1,:,:] = var_2_stationary_mean
    create_netcdf_point_eddy(var_2_overall_pool,var_2_transient_pool,var_2_transient_mean_pool,
                             var_2_standing_pool,var_2_stationary_mean_pool,var_2_steady_mean,output_path)
    print ('The full pipeline of the decomposition of meridional energy transport in the atmosphere is accomplished!')
    print ("--- %s minutes ---" % ((tttt.time() - start_time)/60))
```
